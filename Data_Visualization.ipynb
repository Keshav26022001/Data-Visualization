## Line Chart: 

```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
sales = [100, 200, 300, 400, 500]

plt.plot(days, sales, marker='o', color = 'red')
plt.title('Sales Trend Over Days')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.grid()
plt.show()
```

## Bar Chart

```python
categories = ['Apples', 'Bananas', 'Cherries']
counts = [50, 75, 100]

plt.bar(categories, counts, color='skyblue')
plt.title('Fruit Sales')
plt.xlabel('Fruit')
plt.ylabel('Quantity Sold')
plt.show()
```

## Scatter Plot

```python
age = [25, 30, 35, 40, 45]
income = [30000, 40000, 50000, 60000, 70000]

plt.scatter(age, income, color='red')
plt.title('Income vs Age')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

## Histogram

```python
data = [55, 60, 65, 70, 75, 80, 85, 90]

plt.hist(data, bins=5, color='pink', edgecolor= 'black')
plt.title('Grade Distribution')
plt.xlabel('Grades')
plt.ylabel('Frequency')
plt.show()
```

## Pie Chart 

```python
labels = ['Rent', 'Food', 'Transport', 'Savings']
sizes = [30, 40, 15, 15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Monthly Expenses')
plt.show()
```

## Box Plot

```python
scores = [70, 75, 80, 85, 90, 95, 100]

plt.boxplot(scores, patch_artist=True)
plt.title('Score Distribution')
plt.ylabel('Scores')
plt.show()
```

## Heatmap

```python
import numpy as np

data = np.random.rand(4, 4)

plt.imshow(data, cmap='hot')
plt.colorbar()
plt.title('Heatmap Example')
plt.show()
```

## Area Chart

```python
months = [1, 2, 3, 4, 5]
revenue = [100, 200, 300, 400, 500]

plt.fill_between(months, revenue, color="orange", alpha=0.5)
plt.plot(months, revenue, color="red")
plt.title('Revenue Over Time')
plt.xlabel('Months')
plt.ylabel('Revenue')
plt.show()
```

## 3D Plot

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = [1, 2, 3, 4, 5]
y = [5, 6, 7, 8, 9]
z = [10, 11, 12, 13, 14]

ax.scatter(x, y, z, color='blue', marker='o')
plt.title('3D Data Points')
plt.show()
```

## Stacked Bar Chart

```python
category_A = [3, 6, 8]
category_B = [4, 5, 9]

labels = ['2021', '2022', '2023']

plt.bar(labels, category_A, label='Category A', color='yellow')
plt.bar(labels, category_B, bottom=category_A, label='Category B', color='blue')
plt.title('Sales by Year')
plt.legend()
plt.show()
```

## Step Plot

```python
x = [1, 2, 3, 4]
y = [2, 5, 8, 11]

plt.step(x, y, color='green', where='post')
plt.title('Step Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## Error Bar

```python
x = [1, 2, 3]
y = [2, 4, 6]
errors = [0.1, 0.2, 0.3]

plt.errorbar(x, y, yerr=errors, fmt='o', capsize=3, label="Data with error bars")
plt.title('Error Bars Example')
plt.legend()
plt.show()
```

## Density Plot

```python
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, density=True, alpha=0.5, label='Histogram')
plt.title('Density Plot Example')
plt.xlabel('Values')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for plots
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]  # Plot 1 data
y2 = [1, 2, 3, 4, 5]    # Plot 2 data

# Create a figure with 1 row and 2 columns of subplots
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.plot(x, y1, color='blue', label="Plot 1")
plt.title("First Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()

plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.plot(x, y2, color='green', label="Plot 2")
plt.title("Second Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()

# Display both subplots
plt.tight_layout()  # Adjust spacing to prevent overlap
plt.show()
```
